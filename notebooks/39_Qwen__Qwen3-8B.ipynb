```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="/disk/ssd1/zihaomu_amd/models/Qwen__Qwen3-8B", device_map="auto")
messages = [
    {"role": "user", "content": "Who are you?"},
]
print(pipe(messages, max_new_tokens=64))
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tok = AutoTokenizer.from_pretrained("/disk/ssd1/zihaomu_amd/models/Qwen__Qwen3-8B")
model = AutoModelForCausalLM.from_pretrained("/disk/ssd1/zihaomu_amd/models/Qwen__Qwen3-8B", torch_dtype="auto", device_map="auto")
messages = [{"role": "user", "content": "Who are you?"}]
inputs = tok.apply_chat_template(messages, add_generation_prompt=True, tokenize=True, return_dict=True, return_tensors="pt").to(model.device)
out = model.generate(**inputs, max_new_tokens=64)
print(tok.decode(out[0][inputs["input_ids"].shape[-1]:]))
```
