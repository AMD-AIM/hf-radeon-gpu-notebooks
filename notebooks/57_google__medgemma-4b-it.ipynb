```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("image-text-to-text", model="google/medgemma-4b-it", device_map="auto")
messages = [{'role': 'user',
  'content': [{'type': 'image',
               'url': 'https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/p-blog/candy.JPG'},
              {'type': 'text', 'text': 'What animal is on the candy?'}]}]
print(pipe(messages, max_new_tokens=64))
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tok = AutoTokenizer.from_pretrained("google/medgemma-4b-it")
model = AutoModelForCausalLM.from_pretrained("google/medgemma-4b-it", torch_dtype="auto", device_map="auto")
messages = [{'role': 'user',
  'content': [{'type': 'image',
               'url': 'https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/p-blog/candy.JPG'},
              {'type': 'text', 'text': 'What animal is on the candy?'}]}]
inputs = tok.apply_chat_template(messages, add_generation_prompt=True, tokenize=True, return_dict=True, return_tensors="pt").to(model.device)
out = model.generate(**inputs, max_new_tokens=64)
print(tok.decode(out[0][inputs["input_ids"].shape[-1]:]))
```
