## Local Inference on GPU 
Model page: https://huggingface.co/openbmb/MiniCPM5-1B

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/openbmb/MiniCPM5-1B)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="openbmb/MiniCPM5-1B", device_map="auto")
messages = [
    {"role": "user", "content": "Who are you?"},
]
print(pipe(messages, max_new_tokens=64))
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tok = AutoTokenizer.from_pretrained("openbmb/MiniCPM5-1B")
model = AutoModelForCausalLM.from_pretrained("openbmb/MiniCPM5-1B", torch_dtype="auto", device_map="auto")
messages = [{"role": "user", "content": "Who are you?"}]
inputs = tok.apply_chat_template(messages, add_generation_prompt=True, tokenize=True, return_dict=True, return_tensors="pt").to(model.device)
out = model.generate(**inputs, max_new_tokens=64)
print(tok.decode(out[0][inputs["input_ids"].shape[-1]:]))
```
