```python
# Fixed local validation for deepseek-ai/DeepSeek-OCR-2
# The model remote code expects transformers==4.46.3 because newer transformers removed LlamaFlashAttention2.
# Keep this runtime change local to this notebook; later notebooks can upgrade transformers again.
import subprocess as _nbsp

# ===== code cell =====
_nbsp.run(r"""pip install -U "transformers==4.46.3" accelerate""", shell=True, check=True)

# ===== code cell =====
import torch
from transformers import AutoTokenizer, AutoModel

model_path = 'deepseek-ai/DeepSeek-OCR-2'
print("cuda_available", torch.cuda.is_available())
print("device_count", torch.cuda.device_count())

tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
model = AutoModel.from_pretrained(
    model_path,
    trust_remote_code=True,
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True,
)
print(type(tokenizer).__name__)
print(type(model).__name__)
print("loaded")
```
