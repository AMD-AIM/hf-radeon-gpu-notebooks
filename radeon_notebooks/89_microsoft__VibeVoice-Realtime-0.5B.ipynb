```python
import json
from pathlib import Path
from transformers import AutoConfig

model_path = "microsoft/VibeVoice-Realtime-0.5B"
try:
    cfg = AutoConfig.from_pretrained(model_path, trust_remote_code=True)
    print("config:", type(cfg).__name__, "model_type=", getattr(cfg, "model_type", None))
except Exception as e:
    cfg_json = json.loads((Path(model_path) / "config.json").read_text(encoding="utf-8"))
    print("fallback model_type:", cfg_json.get("model_type"), "load_error:", type(e).__name__)
```
