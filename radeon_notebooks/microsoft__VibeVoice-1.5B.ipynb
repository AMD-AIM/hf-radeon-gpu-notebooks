## Local Inference on GPU 
Model page: https://huggingface.co/microsoft/VibeVoice-1.5B

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/microsoft/VibeVoice-1.5B)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
# Use a pipeline as a high-level helper
import gc
import torch
from transformers import pipeline

model_path = "microsoft/VibeVoice-1.5B"
pipe = pipeline(
    "text-to-speech",
    model=model_path,
    device_map="auto",
    trust_remote_code=True,
)
result = pipe("Hello, this is a short Radeon CI speech generation test.")
print("result keys =", sorted(result.keys()))
print("audio shape =", getattr(result["audio"], "shape", None))
print("sampling_rate =", result.get("sampling_rate"))
pipe_parameter_devices = sorted({str(p.device) for p in pipe.model.parameters()})
print("pipe parameter devices =", pipe_parameter_devices)

del pipe
gc.collect()
torch.cuda.empty_cache()
```

```python
# Load model directly
from transformers import AutoModelForSeq2SeqLM, pipeline

model = AutoModelForSeq2SeqLM.from_pretrained(
    model_path,
    torch_dtype="auto",
    device_map="auto",
    trust_remote_code=True,
).eval()
pipe = pipeline(
    "text-to-speech",
    model=model,
    tokenizer=model_path,
    feature_extractor=model_path,
    trust_remote_code=True,
)
result = pipe("Hello, this is a direct Radeon CI speech generation test.")
print("result keys =", sorted(result.keys()))
print("audio shape =", getattr(result["audio"], "shape", None))
print("sampling_rate =", result.get("sampling_rate"))
model_parameter_devices = sorted({str(p.device) for p in model.parameters()})
print("model parameter devices =", model_parameter_devices)
```
